**SVM Support Vector Machine: Is used for classfication,regression and outilers dection tasks, it works by finding the optimal hyperplane that separate data points of different class in high dimensional space**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
iris = load_iris()
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df["target"] = iris.target
df.head()
```

```python
df[df.target ==1].head()
```

```python
df[df.target==2].head()
```

```python
#getting the flowers names
df["flower_name"] = df.target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
#Train using suport vector machine (SVM)
x = df.drop(["target", "flower_name"], axis = "columns")
y =df.target
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
model = SVC()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
